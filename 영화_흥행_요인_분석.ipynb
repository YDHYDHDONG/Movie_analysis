## 연도별 Meta Score 추이

```python
import plotly.express as px
```

```python
year_metascore = data.groupby('Year')[['Meta Score']].mean().reset_index()
```

```python
fig = px.line(data_frame = year_metascore, x = 'Year', y = 'Meta Score', title = '연도별 Meta Score 추이')
fig.show()
```

```python
data.dtypes
```

## 연도별 영화 시청 등급 추이

```python
p_r = data.groupby('Year')[['PG Rating']].value_counts() # PG Rating 등급 발행 횟수
p_r.info()
```

```python
pr_df = p_r.reset_index(name='PG Rating Count')
pr_df
```

```python
fig = px.bar(data_frame=pr_df, x='Year', y='PG Rating Count', color = 'PG Rating',
             title='연도별 PG Rating별 영화 시청 횟수 추이')
fig.show()
```

## 가장 흥행한 영화 TOP 10


```python
# 투표 TOP 10 영화
best_vote = data.groupby('Movie Name')[['Votes']].sum().reset_index().sort_values('Votes', ascending=False).color=.head(10)
fig = px.bar(data_frame = best_vote, x = 'Movie Name', y = 'Votes', title = '투표 Top 10 영화')
fig.show()
```

```python
# 메타스코어 TOP 10
best_meta = data.groupby('Movie Name')[['Meta Score']].mean().reset_index().sort_values('Meta Score', ascending = False).head(10)
fig = px.line(data_frame = best_meta, x = 'Movie Name', y = 'Meta Score')
fig.show()
```
