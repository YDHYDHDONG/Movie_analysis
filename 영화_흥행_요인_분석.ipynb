## 연도별 Meta Score 추이

```python
import plotly.express as px
```

```python
year_metascore = data.groupby('Year')[['Meta Score']].mean().reset_index()
```

```python
fig = px.line(data_frame = year_metascore, x = 'Year', y = 'Meta Score', title = '연도별 Meta Score 추이')
fig.show()
```

## 연도별 영화 시청 등급 추이

```python
year_pg_rating = data.groupby('Year')[['PG Rating']].reset_index()
fig = px.bar(year_pg_rating, x = 'Year', y = 'PG Rating')
```

## 가장 흥행한 영화 TOP 10

> 인용구 추가



```python
# 투표 TOP 10 영화
best_vote = data.groupby('Movie Name')[['Votes']].sum().reset_index().sort_values('Votes', ascending=False).head(10)
fig = px.bar(data_frame = best_vote, x = 'Movie Name', y = 'Votes', title = '투표 Top 10 영화')
fig.show()
```

```python
# 메타스코어 TOP 10
best_meta = data.groupby('Movie Name')[['Meta Score']].mean().reset_index().sort_values('Meta Score', ascending = False).head(10)
fig = px.line(data_frame = best_meta, x = 'Movie Name', y = 'Meta Score')
fig.show()
```
